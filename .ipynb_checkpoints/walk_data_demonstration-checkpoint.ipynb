```python
n_cls = 3
mfcc_cval_sample = np.zeros((2500, 13, 3))
j_input_cval_sample = np.zeros((2500, 50, 3))

a = np.reshape(np.arange(mfcc.shape[-1]), (3, -1))

for i, j in enumerate(a):
    mfcc_cval_sample[:, :, i] = f.flatten_ax(mfcc_sc[:, :, j])
    j_input_cval_sample[:, :, i] = f.flatten_ax(j_input_sc[:, :, j])

    
plt.plot(j_input_cval_sample[:, 2, 0])
plt.plot(j_input_cval_sample[:, 14, 0])
```

```python
n_folds = 5
nets = []

mfcc_cval_train, mfcc_cval_test = f.make_chunks(mfcc_cval_sample, n_folds)
j_input_cval_train, j_input_cval_test= f.make_chunks(j_input_cval_sample, n_folds)
#preds_cval = np.zeros(np.concatenate((mfcc_cval_test.shape[:-1], [n_cls], [n_folds])))
dists = np.zeros((mfcc_cval_test.shape[0], n_cls, n_cls, n_folds))
conf_mats = np.zeros((n_cls, n_cls, n_folds))
```

```python
for i in range(n_folds):
    print "Fold " + str(i+1)
    nets_fold = []
    for j in range(n_cls):
        print "Net" + str(j+1)
        nets_fold.append(f.learn_mfcc(mfcc_cval_train[:, :, j, i],\
                                      j_input_cval_train[:, :, j, i], epochs = 80, inter_layer = 20))
        for k in range(n_cls):
            preds_cval_temp = f.predict_mfcc(nets_fold[j], j_input_cval_test[:, :, k, i])
            dists[:, j, k, i] = np.linalg.norm(mfcc_cval_test[:, :, k, i] - preds_cval_temp, axis = 1)
            
        preds_class = np.argmin(dists[:, :, :, i], axis = 1)
    conf_mats[:, :, i] = f.calculate_confusion_matrix(preds_class, np.arange(n_cls))
    nets.append(nets_fold)

avg_conf_mat = np.average(conf_mats, axis = 2)

print "Averaged confusion matrix:\n", avg_conf_mat
```
